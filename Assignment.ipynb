```python
z=int(input("Enter the number of students"))
name=[]
mark=[]
gradescal=[]
avg=0
sum=0
passed=0
fail=0

def grade(mark):
  if mark>=90:
    gradescal.append("A")
  elif mark>=80:
    gradescal.append("B")
  elif mark>=70:
    gradescal.append("C")
  elif mark>=60:
    gradescal.append("D")
  else:
    gradescal.append("F")



for i in range(0,z):
  print("Enter the name of student")
  name.append(input())
  print("Enter the marks of student")
  mark.append(int(input()))
  sum=sum+mark[i]
  grade(mark[i])
avg=sum/z

print("Student Report")
print("--------------")
for i in range(0,z):
  print(f"{name[i]}:{mark[i]}:{gradescal[i]}")

for i in range(0,z):
  if gradescal[i]=="F":
    fail=fail+1
  else:
    passed=passed+1


print(f"\nStudents Passed:{passed}")
print(f"Students Failed:{fail}")

print(f"Class Average:{avg}")
if avg>=85:
  print(f"Overall Performance: Excellent")
elif avg>=70:
  print(f"Overall Performance: Good")
elif avg>=50:
  print(f"Overall Performance: Average")
elif avg<50:
  print(f"Overall Performance: Needs Improvement")







```
