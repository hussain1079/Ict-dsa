**Relational Operators**

```python
costpen=10
costcar=250
print(costpen>costcar)
print(costpen<costcar)
print(costpen==costcar)
print(costpen!=costcar)
print(costpen>=costcar)
print(costpen<=costcar)
```

**Logical Operators**

```python
costpen=10
costcar=250
print(costpen<100 and costcar>costpen)
print(costpen<100 or costcar<100)
print(costpen>100 or costcar<100)
print(not costpen>50)


```
